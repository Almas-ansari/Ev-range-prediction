```python
plt.figure(figsize=(20, 10))
plt.suptitle("transformation")
plt.subplot(1, 3, 1)
plt.title("square")
sns.distplot(np.square(data['consumption(kWh/100km)']), kde = True)

plt.subplot(1, 3, 2)
plt.title("Cube root")
sns.distplot(np.cbrt(data['consumption(kWh/100km)']))

plt.subplot(1,3,3)
plt.title("lograthmic")
sns.distplot(np.log(data['consumption(kWh/100km)']))

plt.show()
```
