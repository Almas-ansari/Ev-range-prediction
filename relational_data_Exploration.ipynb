```python
np.sort(data['avg_speed(km/h)'].unique())
```

```python
def transform(a):
    atrs=[]
    for i in a:
        if i<30:
            i=20
        elif i < 50:
            i=30
        elif i<70:
            i=60
        elif i <100:
            i=80
        atrs.append(i)
    return atrs
```

```python
trs_avg=transform(avg)
```

```python
print(trs_avg)
```

```python
figure(figsize=(14, 8), dpi=80)
sns.scatterplot(X,Y,hue=trs_avg,style=drs,size=tire,palette='magma')
plt.xlabel("fuel")
plt.ylabel("distance")
plt.show()
```
